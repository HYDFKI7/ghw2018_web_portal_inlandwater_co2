```python
import pandas as pd
import numpy as np
```

```python
# read in the co2 dataset
dataset = pd.read_excel("/Users/lsd/Documents/postdoc_proj/co2 measurements.xlsx")

# creating a paperlist
# extract all paper names in the dataset
# rm all non-alnum characters from paper names
# store original and new paper names as a dataframe
paperlist = dataset['Reference'].unique()
cols = ['OrigTitle', 'NewTitle']
paper_name_pairs = []
for paper_name in paperlist:
    new_paper_name = ''.join(string for string in paper_name if string.isalnum())
    paper_name_pairs.append([paper_name, new_paper_name])
papers = pd.DataFrame(paper_name_pairs, columns = cols)
```

```python
# read in the template for co2 datasets
# column names of the template will be used to be create a new formated dataframe for dataset_specpaper
datatemplate = pd.read_excel("/Users/lsd/Documents/ghw2018_web_portal_inlandwater_co2/docs/Data_Template.xlsx",\
                             skiprows = list(range(0,2)))
```

```python
def formatting_subdataset(dataset_specpaper):
    # interate through columns in the dataframe
    # drop the column if it is empty
    for column in dataset_specpaper:
        if dataset_specpaper[column].dropna().empty:
            dataset_specpaper = dataset_specpaper.drop([column], axis = 1)
    
    # create a formated dataframe of dataset_specpaper.shape[0] + 1 in length, with the template's columns
    # row 0 will be units
    dataset_specpaper_formated = \
    pd.DataFrame('NaN', index = range(dataset_specpaper.shape[0] + 1), columns = datatemplate.columns)
    
    # writing 'Date' in dataset_specpaper to 'Sampling Date' in dataset_specpaper_formated
    if ('Date' in dataset_specpaper.columns):
        dataset_specpaper_formated.loc[0,'Sampling Date'] = 'yyyy-mm-dd hh:mm:ss'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Sampling Date'] = list(dataset_specpaper['Date'])

    # writing 'Site_Type' in dataset_specpaper to 'Site Type' in dataset_specpaper_formated
    if 'Site_Type' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Site Type'] = list(dataset_specpaper['Site_Type'])

    # writing 'Lat' or 'Lat_est' in dataset_specpaper to 'Latitude' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Latitude'] = 'decimal degree'
    if 'Lat' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Latitude'] = list(dataset_specpaper['Lat'])
    elif 'Lat_est' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Latitude'] = list(dataset_specpaper['Lat_est'])

    # writing 'Long' or "Long_est" in dataset_specpaper to 'Longitude' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Longitude'] = 'decimal degree'
    if 'Long' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Longitude'] = list(dataset_specpaper['Long'])
    elif 'Long_est' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Longitude'] = list(dataset_specpaper['Long_est'])

    # writing 'Elevation_m' in dataset_specpaper to 'Altitude' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Altitude'] = 'm'
    if 'Elevation_m' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Altitude'] = list(dataset_specpaper['Elevation_m'])

    # writing 'Width_m' in dataset_specpaper to 'Width' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Width'] = 'm'
    if 'Width_m' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Width'] = list(dataset_specpaper['Width_m'])

    # writing 'Depth_m' in dataset_specpaper to 'Depth' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Depth'] = 'm'
    if 'Depth_m' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Depth'] = list(dataset_specpaper['Depth_m'])

    # writing 'surface area (lake, km2)' in dataset_specpaper to 'Surface Area' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Surface Area'] = 'km2'
    if 'surface area (lake, km2)' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Surface Area'] = list(dataset_specpaper['surface area (lake, km2)'])

    # writing 'Discharge_m3s' or 'Discharge_Ls' in dataset_specpaper to 'Discharge' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Discharge'] = 'm3 s-1'
    if 'Discharge_m3s' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Surface Area'] = list(dataset_specpaper['Discharge_m3s'])
    elif 'Discharge_Ls' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'Discharge'] = 'L s-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Surface Area'] = list(dataset_specpaper['Discharge_Ls'])

    # writing 'Water_Velocity_ms' in dataset_specpaper to 'Flow Velocity' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Flow Velocity'] = 'm s-1'
    if 'Water_Velocity_ms' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Flow Velocity'] = list(dataset_specpaper['Water_Velocity_ms'])

    # ! writing 'Wind_Speed_ms' in dataset_specpaper to 'Wind Speed' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Wind Speed'] = 'm s-1'
    if 'Wind_Speed_ms' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Wind Speed'] = list(dataset_specpaper['Wind_Speed_ms'])

    # writing 'Tw_degc' or 'Tw_degc_est' in dataset_specpaper to 'Water Temperature' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Water Temperature'] = '°C'
    if 'Tw_degc' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Water Temperature'] = list(dataset_specpaper['Tw_degc'])
    elif 'Tw_degc_est' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Water Temperature'] = list(dataset_specpaper['Tw_degc_est'])

    # writing 'Air_T_degc' in dataset_specpaper to 'Air Temperature' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Air Temperature'] = '°C'
    if 'Air_T_degc' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Air Temperature'] = list(dataset_specpaper['Air_T_degc'])

    # writing 'CO2_uatm', 'CO2_ppmv', 'CO2_umoll', 'CO2_mgl', 'CO2_mgCl', 'excess_CO2_uM', 'EpCO2', 'fugacity_CO2_uatm' 
    #\in dataset_specpaper to 'CO2' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'CO2'] = 'μatm'
    if 'CO2_uatm' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['CO2_uatm'])
    elif 'CO2_ppmv' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2'] = 'ppm'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['CO2_ppmv'])
    elif 'CO2_umoll' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2'] = 'μmol L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['CO2_umoll'])
    elif 'CO2_mgl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2'] = 'mg L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['CO2_mgl'])
    elif 'CO2_mgCl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2'] = 'mg C L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['CO2_mgCl'])
    elif 'excess_CO2_uM' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2'] = 'μM'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['excess_CO2_uM'])
    elif 'EpCO2' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2'] = 'μatm'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['EpCO2'])
    elif 'fugacity_CO2_uatm' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2'] = 'μatm'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2'] = list(dataset_specpaper['fugacity_CO2_uatm'])

    # writing 'CH4_umoll', 'ECH4_umoll', 'CH4_nmoll', 'CH4_ugl', 'CH4_ugCl', 'CH4_uatm', 'CH4_ppm'
    #\in dataset_specpaper to 'CH4' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'CH4'] = 'μatm'
    if 'CH4_uatm' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CH4'] = list(dataset_specpaper['CH4_uatm'])
    elif 'CH4_umoll' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CH4'] = 'μmol L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CH4'] = list(dataset_specpaper['CH4_umoll'])
    elif 'ECH4_umoll' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CH4'] = 'μmol L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CH4'] = list(dataset_specpaper['ECH4_umoll'])
    elif 'CH4_nmoll' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CH4'] = 'nmol L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CH4'] = list(dataset_specpaper['CH4_nmoll'])
    elif 'CH4_ugl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CH4'] = 'μg L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CH4'] = list(dataset_specpaper['CH4_ugl'])
    elif 'CH4_ugCl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CH4'] = 'μg L-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CH4'] = list(dataset_specpaper['CH4_ugCl'])
    elif 'CH4_ppm' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CH4'] = 'ppm'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CH4'] = list(dataset_specpaper['CH4_ppm'])

    # writing 'N2O_nmoll' in dataset_specpaper to 'N2O' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'N2O'] = 'nmol L-1'
    if 'N2O_nmoll' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'N2O'] = list(dataset_specpaper['N2O_nmoll'])

    # writing 'CO2_Flux_gCm2yr', 'CO2_Flux_mgCm2d', 'CO2_Flux_mgCm2hr', 'CO2_Flux_mgChayr', 
    #\'CO2_Flux_ugCm2s', 'CO2_Flux_gm2d', 'CO2_Flux_umolm2s',
    #\'CO2_Flux_mmolm2d', 'CO2_Flux_molm2d', 'CO2_Flux_umolmh', in dataset_specpaper 
    #\to 'CO2 Flux' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'g C m-2 yr-1'
    if 'CO2_Flux_gCm2yr' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_gCm2yr'])
    elif 'CO2_Flux_mgCm2d' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'mg C m-2 d-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_mgCm2d'])    
    elif 'CO2_Flux_mgCm2hr' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'mg C m-2 hr-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_mgCm2hr'])    
    elif 'CO2_Flux_mgChayr' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'mg C ha-1 yr-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_mgChayr'])    
    elif 'CO2_Flux_ugCm2s' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'μg C m-2 s-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_ugCm2s'])    
    elif 'CO2_Flux_gm2d' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'g C m-2 d-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_gm2d'])    
    elif 'CO2_Flux_umolm2s' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'μmol m-2 s-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_umolm2s'])    
    elif 'CO2_Flux_mmolm2d' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'mmol m-2 d-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_mmolm2d'])    
    elif 'CO2_Flux_molm2d' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'mol m-2 d-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_molm2d'])    
    elif 'CO2_Flux_umolmh' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'CO2 Flux'] = 'μmol m-2 h-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'CO2 Flux'] = list(dataset_specpaper['CO2_Flux_umolmh'])    
   
    # writing 'k_md', 'k_cmh' in dataset_specpaper to 'k' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'k'] = 'm d-1'
    if 'k_md' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'k'] = list(dataset_specpaper['k_md'])
    elif 'k_cmh' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'k'] = 'cm h-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'k'] = list(dataset_specpaper['k_cmh'])

    # writing 'k600_md', 'k600_cmh' in dataset_specpaper to 'k600' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'k600'] = 'm d-1'
    if 'k600_md' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'k600'] = list(dataset_specpaper['k600_md'])
    elif 'k600_cmh' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'k600'] = 'cm h-1'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'k600'] = list(dataset_specpaper['k600_cmh'])

    # writing 'DOC_mgl' in dataset_specpaper to 'DOC' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'DOC'] = 'mg L-1'
    if 'DOC_mgl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'DOC'] = list(dataset_specpaper['DOC_mgl'])

    # writing 'POC_mgl' in dataset_specpaper to 'POC' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'POC'] = 'mg L-1'
    if 'POC_mgl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'POC'] = list(dataset_specpaper['POC_mgl'])

    # writing 'TOC_mgl' in dataset_specpaper to 'TOC' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'TOC'] = 'mg L-1'
    if 'TOC_mgl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'TOC'] = list(dataset_specpaper['TOC_mgl'])

    # writing 'pH' in dataset_specpaper to 'pH' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'pH'] = 'NaN'
    if 'pH' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'pH'] = list(dataset_specpaper['pH'])

    # writing 'alk_uM', 'Total_DIC_umolkg','Total_DIC_mgl' in dataset_specpaper
    #\to 'Alkalinity' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Alkalinity'] = 'mg L-1'
    if 'Total_DIC_mgl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Alkalinity'] = list(dataset_specpaper['Total_DIC_mgl'])
    elif 'alk_uM' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'Alkalinity'] = 'μM'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Alkalinity'] = list(dataset_specpaper['alk_uM'])
    elif 'Total_DIC_umolkg' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'Alkalinity'] = 'μM'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Alkalinity'] = list(dataset_specpaper['Total_DIC_umolkg'])

    # writing 'O2_mgl', 'O2_per' in dataset_specpaper to 'DO' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'DO'] = 'mg L-1'
    if 'O2_mgl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'DO'] = list(dataset_specpaper['O2_mgl'])
    elif 'O2_per' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[0,'DO'] = '%'
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'DO'] = list(dataset_specpaper['O2_per'])

    # writing 'Chl_a_ugl' in dataset_specpaper to 'Chl a' in dataset_specpaper_formated
    dataset_specpaper_formated.loc[0,'Chl a'] = 'μg L-1'
    if 'Chl_a_ugl' in dataset_specpaper.columns:
        dataset_specpaper_formated.loc[1:dataset_specpaper.shape[0],'Chl a'] = list(dataset_specpaper['Chl_a_ugl'])
    
    # return the subset
    return dataset_specpaper_formated
```

```python
def write_to_excel(dataset_specpaper_formated):
    writer = pd.ExcelWriter('/Users/lsd/Documents/postdoc_proj/excel/' + papers.loc[i, 'NewTitle'] + '.xlsx')
    dataset_specpaper_formated.to_excel(writer, 'Sheet1')
    writer.save()
```

```python
# The main program for formatting sub-datasets
for i in range(0,len(papers)):
    dataset_specpaper = dataset.loc[dataset['Reference'] == papers['OrigTitle'][i],]
    dataset_specpaper_formated = formatting_subdataset(dataset_specpaper)
    write_to_excel(dataset_specpaper_formated)
    print("Formatting paper number {} with the title of '{}'.".format(i, papers.loc[i,'NewTitle']))
```
